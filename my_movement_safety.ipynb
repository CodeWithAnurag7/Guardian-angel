```python
# Install required libraries if not already
!pip install folium geopy

import folium
from geopy.distance import geodesic
import time

# ------------------------------
# Setup: Initial tourist location & danger zones
# ------------------------------
tourist_location = (28.6139, 77.2090)  # Example: Connaught Place, Delhi
danger_zones = [
    {"name": "Riot Area", "location": (28.6200, 77.2100), "radius": 600},
    {"name": "Flooded Area", "location": (28.6000, 77.2000), "radius": 700}
]

SAFETY_RADIUS = 500  # tourist bubble in meters

# ------------------------------
# Function to check danger
# ------------------------------
def check_danger(tourist_loc):
    for zone in danger_zones:
        dist = geodesic(tourist_loc, zone["location"]).meters
        if dist <= SAFETY_RADIUS + zone["radius"]:
            return True, zone["name"]
    return False, None

# ------------------------------
# Create map function
# ------------------------------
def create_map(tourist_loc):
    m = folium.Map(location=tourist_loc, zoom_start=14)

    # Tourist bubble
    folium.Circle(
        radius=SAFETY_RADIUS,
        location=tourist_loc,
        color="green",
        fill=True,
        fill_opacity=0.3,
        popup="Tourist Safety Bubble"
    ).add_to(m)

    # Danger zones
    for zone in danger_zones:
        folium.Circle(
            radius=zone["radius"],
            location=zone["location"],
            color="red",
            fill=True,
            fill_opacity=0.4,
            popup=f"Danger: {zone['name']}"
        ).add_to(m)

    # Check if danger is near
    danger, reason = check_danger(tourist_loc)
    if danger:
        folium.Marker(
            location=tourist_loc,
            popup=f"⚠️ ALERT: You are entering danger zone ({reason})!",
            icon=folium.Icon(color="red")
        ).add_to(m)
    else:
        folium.Marker(
            location=tourist_loc,
            popup="✅ Safe Area",
            icon=folium.Icon(color="green")
        ).add_to(m)

    return m

# ------------------------------
# Simulate tourist movement
# ------------------------------
path = [
    (28.6139, 77.2090),  # Start (safe)
    (28.6170, 77.2105),  # Moving closer
    (28.6190, 77.2107),  # Entering danger zone
]

maps = []
for loc in path:
    maps.append(create_map(loc))

# Show last map (final tourist position)
maps[-1]
```
